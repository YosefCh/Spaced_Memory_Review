## Visual 2 — Weekly Learning Consistency
A line chart showing how many days per 7-day block had at least one learning entry.  
- **Green dashed line** = perfect week (7/7 days)  
- **Orange dashed line** = your personal goal — use the slider below to adjust it  

Useful for spotting streaks, dips, and overall consistency trends over time.


```python
from ipywidgets import interact, IntSlider
interact(
    plot_weekly_consistency,
    goal=IntSlider(value=5, min=1, max=7, step=1, description='Goal (days):')
);
```

---
## Visual 3 — Gap Analysis
A horizontal bar chart showing how many days have passed since each subject was last studied.  
- **Green** = studied more recently than the yellow threshold  
- **Yellow** = between the yellow and red thresholds  
- **Red** = beyond the red threshold  

Use the sliders below to set your own threshold values and the chart will update live.


```python
from ipywidgets import interact, IntSlider
interact(
    plot_gap_analysis,
    yellow_threshold=IntSlider(value=150, min=7,  max=730, step=7, description='Yellow (days):'),
    red_threshold=   IntSlider(value=300, min=14, max=730, step=7, description='Red (days):'  )
);
```

## Visual 4 — Learning Days vs. Empty Days
A pie chart showing what percentage of days in a selected period had at least one valid learning entry.  
A day only counts as "learned" if its **Subject** or **Topic** field is non-empty.  
Use the date pickers below to focus on any time window — defaults to your full history.


```python
import datetime
import pandas as pd
from ipywidgets import interact, DatePicker

# Determine the actual date range in the data to use as defaults
_df = pd.read_csv(r'C:\Users\Rebecca\OneDrive\Documents\Review\Data\learned_material.csv')
_df['Date'] = pd.to_datetime(_df['Date'])
_data_start = _df['Date'].min().date()
_data_end   = datetime.date.today()

interact(
    plot_learning_pie,
    start_date=DatePicker(value=_data_start, description='Start date:'),
    end_date=  DatePicker(value=_data_end,   description='End date:')
);
```
